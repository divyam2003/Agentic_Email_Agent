````python
import re

def parse_enron_email_tmp(raw_email):
    # Extract Subject
    subject_match = re.search(r"Subject:(.*)", raw_email)
    subject = subject_match.group(1).strip() if subject_match else ""
    # Extract Body (after first blank line)
    parts = raw_email.split("\n\n",1)

    # print(parts)
    body = parts[1] if len(parts) > 1 else ""
    # Remove extra whitespace
    body = re.sub(r"\s+", " ", body).strip()

    # -----------------------------------
    body = "\n".join(
        line for line in body.split("\n")
        if not line.strip().startswith(">")
    )
    # Remove forwarded/reply chains
    body = re.split(r"-----Original Message-----", body)[0]
    body = re.split(r"----------- Forwarded by", body)[0]

    # Remove signatures (basic)
    body = re.split(r"(Regards|Thanks|Sincerely)", body)[0]
    # Final cleanup
    body = re.sub(r"\s+", " ", body).strip()
    part = raw_email.split("\n")
    # print(part)
    metadata=''
    for i in range (1,len(part)):
        if(len(part[i]) > 1 and part[i]!=''):
            metadata = metadata + part[i]

    return subject, body, metadata
````

````python
import re

def parse_enron_email(raw_email):
    # print(raw_email)
    # print(";;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;")
    # Extract Subject
    subject_match = re.search(r"Subject:(.*)", raw_email)
    # print("subject_match--->",subject_match)
    subject = subject_match.group(1).strip() if subject_match else ""
    # print("subject--->",subject)
    # Extract Body (after first blank line)
    parts = raw_email.split("\n\n", 1)
    # print("parts--->",parts)
    body = parts[1] if len(parts) > 1 else ""
    # print("body---->  ------->",body)
    # Remove extra whitespace
    body = re.sub(r"\s+", " ", body).strip()
    # print("body---->",body)
    # print()
    # print("//////////////////////////////////////////////")


    return subject, body
````

````python
def clean_body(body):
    # Remove quoted replies
    # print("clean_body -........................................")
    # print("body---->  ------->",body)

    body = "\n".join(
        line for line in body.split("\n")
        if not line.strip().startswith(">")
    )
    # print("body---->  ------->",body)

    # Remove forwarded/reply chains
    body = re.split(r"-----Original Message-----", body)[0]
    # print("body---->  ------->",body)

    # Remove signatures (basic)
    body = re.split(r"(Regards|Thanks|Sincerely)", body)[0]
    # print("body---->  ------->",body)

    # Final cleanup
    body = re.sub(r"\s+", " ", body).strip()
    # print("body---->  ------->",body)

    return body
````

````python
def to_training_sample(subject, body, metadata=None):
    # email_text = f"{subject}\n\n{body}" if subject else body
    # email_text = f"{subject}\n{metadata}\n\n{body}" if subject and metadata else body
    email_text = "\n".join(filter(None, [subject, metadata, "", body]))

    return email_text
````

````python
def process_email(raw_email):
    # subject, body = parse_enron_email(raw_email)
    # body = clean_body(body)
    subject, body,metadata = parse_enron_email_tmp(raw_email)

    prompt = to_training_sample(subject, body)

    return prompt
````

````python
import json
import re

def clean_json_output(text):
    # Remove ```json ``` wrappers if present
    text = re.sub(r"```json\s*", "", text)
    text = re.sub(r"```", "", text)

    # Strip whitespace
    text = text.strip()

    # Try to extract JSON block if extra text exists
    try:
        # Find first { ... } block
        match = re.search(r"\{.*\}", text, re.DOTALL)
        if match:
            text = match.group(0)

        return json.loads(text)
    except Exception as e:
        return None


def convert_row(row):
    system_prompt = "You are an email assistant. Extract structured information and return ONLY valid JSON."

    user_prompt = f"Analyze the email:\n\n{row[1]}"

    cleaned = clean_json_output(row[2])

    if cleaned is None:
        return None  # skip bad rows

    return {
        "messages": [
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt},
            {"role": "assistant", "content": json.dumps(cleaned, indent=2)}
        ]
    }
````

````python


def extract_email_content(user_text):
    # Extract only the email part after the prompt
    parts = user_text.split("Analyze the email:\n\n", 1)
    return parts[1] if len(parts) > 1 else user_text


def normalize_text(text):
    text = text.lower()
    text = re.sub(r"re:|fw:", "", text)   # optional cleanup
    text = re.sub(r"\s+", " ", text)
    return text.strip()


def remove_duplicates(input_file, output_file):
    seen = set()

    with open(input_file, "r") as fin, open(output_file, "w") as fout:
        for line in fin:
            try:
                data = json.loads(line)

                # get user message
                user_msg = next(
                    msg["content"] for msg in data["messages"] if msg["role"] == "user"
                )

                # extract actual email
                email_text = extract_email_content(user_msg)

                key = normalize_text(email_text)

                if key in seen:
                    continue  # ❌ drop duplicate

                seen.add(key)
                fout.write(json.dumps(data) + "\n")

            except Exception:
                continue
````

````python
import json

def fix_structure(input_file, output_file):
    with open(input_file, "r") as fin, open(output_file, "w") as fout:
        for line in fin:
            try:
                data = json.loads(line)

                # Case 1: [[{...}]]
                if isinstance(data, list) and len(data) > 0:
                    inner = data[0]

                    # If inner is already list of messages
                    if isinstance(inner, list):
                        fixed = {"messages": inner}

                    else:
                        continue
                else:
                    continue

                fout.write(json.dumps(fixed) + "\n")

            except Exception:
                continue
````

````python
import torch
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    BitsAndBytesConfig,
    TrainingArguments
)
from peft import LoraConfig, get_peft_model
from datasets import Dataset
from trl import SFTTrainer
````

````python
# -----------------------
# Model
# -----------------------
model_id = "microsoft/Phi-3.5-mini-instruct"

tokenizer = AutoTokenizer.from_pretrained(model_id)

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4"
)

# model = AutoModelForCausalLM.from_pretrained(
#     model_id,
#     quantization_config=bnb_config,
#     device_map="auto",
#     offload_folder="offload",
#     offload_state_dict=True
# )
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb_config,
    device_map="auto",
    torch_dtype=torch.float16,   # ✅ force FP16
    offload_folder="offload",
    offload_state_dict=True
)

# IMPORTANT memory fixes
model.config.use_cache = False
model.gradient_checkpointing_enable()
````

````python
# -----------------------
# LoRA (PEFT)
# -----------------------
lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=["qkv_proj", "o_proj"],  # ✅ correct for Phi
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM"
)

model = get_peft_model(model, lora_config)
````

````python
# data = [...]  # your 750 samples

def format_example(example):
    text = ""
    for msg in example["messages"]:
        text += f"{msg['role']}: {msg['content']}\n"
    return {"text": text}

dataset = Dataset.from_list(data)
dataset = dataset.map(format_example)
````

````python
def tokenize(example):
    return tokenizer(
        example["text"],
        truncation=True,
        max_length=256,   # keep this low for Kaggle
        padding="max_length"
    )

dataset = dataset.map(tokenize, batched=True)
````

````python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./phi-email-agent",
    per_device_train_batch_size=1,
    gradient_accumulation_steps=8,
    learning_rate=2e-4,
    num_train_epochs=3,
    logging_steps=10,
    save_strategy="epoch",
    fp16=False,
    bf16=False,
    report_to="none",
    optim="adamw_torch"   # ✅ IMPORTANT (avoids fused/AMP paths)
)
````

````python
from trl import SFTTrainer
from transformers import DataCollatorForLanguageModeling


data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False
)

trainer = SFTTrainer(
    model=model,
    train_dataset=dataset,
    data_collator=data_collator,
    args=training_args
)

trainer.train()
````

````python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
from peft import PeftModel

model_id = "microsoft/Phi-3.5-mini-instruct"

tokenizer = AutoTokenizer.from_pretrained(model_id)

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4"
)

base_model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb_config,
    device_map="auto"
)

model = PeftModel.from_pretrained(
    base_model,
    "/content/phi-email-agent-lora"
)

model.eval()
````

````python
prompt = tokenizer.apply_chat_template(
    [
        {
            'role': 'system',
            'content': 'You are an email assistant. Extract structured information and return ONLY valid JSON.'
        },
        {
            'role': 'user',
            'content': "Analyze the email:\n\nRE: Cinergy vs COMED phy spread\nsounds great to me. fletch"
        }
    ],
    tokenize=False,
    add_generation_prompt=True
)
````
